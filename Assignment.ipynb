```python
x=data[['LotArea']]

y=data['SalePrice']

from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=102)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(xtrain,ytrain)
```

```python
linearpred = lr.predict(xtest)
```

```python
from sklearn import metrics
```

```python
print(metrics.mean_squared_error(linearpred,ytest))
```

```python
plt.scatter(ytest,linearpred)
```

```python
Model_1 = LinearRegression()
x1=data[['LotFrontage','LotArea']]
y1=data['SalePrice']

x1train,x1test,y1train,y1test = train_test_split(x1,y1,test_size=0.3,random_state=102)

Model_1.fit(x1train,y1train)
Model_1pred= Model_1.predict(x1test)
print("Coefficients are {}".format(Model_1.coef_))
print('Intercept is {}'.format(Model_1.intercept_))
print('Mean Squared Error {}'.format(metrics.mean_squared_error(y1test,Model_1pred)))
print('R2 Square {}'.format(metrics.r2_score(y1test,Model_1pred)))
```

```python
Model_2 = LinearRegression()
x2=data[['LotFrontage','LotArea','OverallQual', 'OverallCond']]
y2=data['SalePrice']

x2train,x2test,y2train,y2test = train_test_split(x2,y2,test_size=0.3,random_state=102)

Model_2.fit(x2train,y2train)
Model_2pred= Model_2.predict(x2test)
print("Coefficients are {}".format(Model_2.coef_))
print('Intercept is {}'.format(Model_2.intercept_))
print('Mean Squared Error {}'.format(metrics.mean_squared_error(y2test,Model_2pred)))
print('R2 Square {}'.format(metrics.r2_score(y2test,Model_2pred)))
```

```python
Model_3 = LinearRegression()
x3=data[['LotFrontage','LotArea','OverallQual', 'OverallCond','1stFlrSF', 'GrLivArea']]
y3=data['SalePrice']

x3train,x3test,y3train,y3test = train_test_split(x3,y3,test_size=0.3,random_state=102)

Model_3.fit(x3train,y3train)
Model_3pred= Model_3.predict(x3test)
print("Coefficients are {}".format(Model_3.coef_))
print('Intercept is {}'.format(Model_3.intercept_))
print('Mean Squared Error {}'.format(metrics.mean_squared_error(y3test,Model_3pred)))
print('R2 Square {}'.format(metrics.r2_score(y3test,Model_3pred)))
```

----> From above three models by comparing r2_score we can simply notice that by increasing the training features the model accuracy or model predictions also increasing 

Note : R2_score lies in range of 0 and 1.. lower the r2_score lower is the value dependent on that features.

```python
from sklearn.preprocessing import PolynomialFeatures
xp = data[['LotArea']]
yp = data['SalePrice']

xptrain,xptest,yptrain,yptest = train_test_split(xp,yp,test_size=0.3,random_state=102)

poly_xp = PolynomialFeatures(degree=2)

xp_train_poly = poly_xp.fit_transform(xptrain)
xp_test_poly = poly_xp.fit_transform(xptest)

Model_Poly = LinearRegression()

Model_Poly.fit(xp_train_poly,yptrain)

Poly_model_pred = Model_Poly.predict(xp_test_poly)

print('Coefficients are {}'.format(Model_Poly.coef_))
print('Intercept is {}'.format(Model_Poly.intercept_))
print('Mean Squared Error {}'.format(metrics.mean_squared_error(yptest,Poly_model_pred)))
print('R2 Score is {}'.format(metrics.r2_score(yptest,Poly_model_pred)))
```
